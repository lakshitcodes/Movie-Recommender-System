```python
# Budget is not a criterion
# Genre is important for recommendation
# Homepage does not makes sense
# Id is required for data identification
# keywords are required, acts as tags
# original_language does not makes sense
# Instead of original_title , we will use title
# Overview gives the similarity , required
# Popularity can be avoided
# production_companies not required
# production_countries not required
# release_date not fitting in our workflow
# revenue not an indicator
# duration does not matter
# spoken_languages does not matter
# status does not matter
# tagline is not a real indicator , is vague
# title is there
# vote_average,vote_count skipping them
# cast is required
# crew (director) is required



# Final columns to keep
# genre
# id 
# keywords
# overview
# title
# cast
# crew
```

```python
movie_dataset = movie_dataset.rename(columns={"title_x" : "title"})
```

```python
movie_dataset = movie_dataset[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movie_dataset.head()
```

```python
movie_dataset.isnull().sum()
```

```python
movie_dataset.dropna(inplace=True)
```

```python
movie_dataset.duplicated().sum()
```

```python
movie_dataset.iloc[0].genres
```

```python
# Convert this to 
# ['Action','Adventure','Fantasy','Science Fiction']
```

```python
import ast
def convert(obj):
    obj = ast.literal_eval(obj)
    genreList = []
    for i in obj:
        genreList.append(i['name'])

    return genreList
```

```python
# genres is present as an string and not as a list of dicts 
# converting usinga module
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# Now will use along with convert method
```

```python
movie_dataset['genres'] = movie_dataset['genres'].apply(convert)
```

```python
movie_dataset.head()
```

```python
# Same on keywords
```

```python
movie_dataset['keywords'] = movie_dataset['keywords'].apply(convert)
```

```python
movie_dataset.head()
```

```python
# Getting top three actors from cast
import ast
def getThreeActors(obj):
    actors = []
    obj = ast.literal_eval(obj)
    for i in range(min(3, len(obj))):
        actors.append(obj[i]['name'])
    return actors
```

```python
movie_dataset['cast'] = movie_dataset['cast'].apply(getThreeActors)
```

```python
movie_dataset.head()
```

```python
movie_dataset['crew'].head(1)
```

```python
# Getting director name

import ast
def getDirector(obj):
    obj = ast.literal_eval(obj)
    director = []
    for i in obj:
        if i['job']=='Director':
            director.append(i['name'])
            break
    return director
```

```python
movie_dataset['crew'] = movie_dataset['crew'].apply(getDirector)
```

```python
movie_dataset.head()
```

```python
# Overview is a string , converting it into a list
```

```python
movie_dataset['overview'] = movie_dataset['overview'].apply(lambda x:x.split())
```

```python
movie_dataset.head()
```

```python
# Removing space from a single name or genre
```

```python
movie_dataset['genres'] = movie_dataset['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie_dataset['keywords'] = movie_dataset['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie_dataset['cast'] = movie_dataset['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie_dataset['crew'] = movie_dataset['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_dataset.head()
```

```python
# Creating a new column tag with conctenatoin of overview, genres, keywords, cast, crew
```

```python
movie_dataset['tags'] = movie_dataset['overview'] + movie_dataset['genres'] + movie_dataset['keywords'] + movie_dataset['cast'] + movie_dataset['crew']
```

```python
movie_dataset.head()
```

```python
# Removing the unwanted columns now
new_df = movie_dataset[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# Converting the tags into a string again
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```
